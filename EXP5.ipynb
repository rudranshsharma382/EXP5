```python
# Dictionary to store student names and marks
student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 65,
    "Eve": 95
}

# Get student name from user input
search_name = input("Enter the student's name to search for marks: ")

# Search for the student's marks
if search_name in student_marks:
    marks = student_marks[search_name]
    print(f"The marks for {search_name} are: {marks}")
else:
    print(f"Student '{search_name}' not found in the record.")
```

```python
# Dictionary to store student names and marks
student_marks = {"Alice": 85,"Bob": 92,"Charlie": 78,"David": 65,"Eve": 95,"Frank": 88}

highest_scorer = max(student_marks, key=student_marks.get)
max_marks = student_marks[highest_scorer]
print(f"The student with the highest marks is {highest_scorer} with {max_marks} marks.")
```
